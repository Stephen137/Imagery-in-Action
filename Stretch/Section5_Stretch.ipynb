Using the skills that you gained in this exercise, you will edit the existing raster function template to include an additional raster function to create the desired output by removing all the pixel values that do not indicate potential ferrous mineral locations. 

1. Go to My Content and open the Possible Ferrous Mineral Locations raster function template in the Raster Function Editor.

2. Search and add the Mask raster function to the Raster Function Template.

3. Connect the output from Boolean And to Mask.

The complete Raster Function Template should look like the following graphic.



4. Review the ArcGIS Online Help: [Mask function](https://doc.arcgis.com/en/arcgis-online/analyze/mask-function.htm) documentation to `determine and apply the appropriate parameters` for the Mask raster function.



5. Save and close the Raster Function Editor.

6. Open the Possible Ferrous Mineral Locations raster function template in the `Fossil Exploration` web map, and then `run the raster function` 



The output will only display the pixel values of potential ferrous mineral locations.
